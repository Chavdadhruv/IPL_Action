```python
import random
import matplotlib.pyplot as plt

class Trade:
    def __init__(self, teams, companies):
        self.teams = teams
        self.companies = companies

    def buy_team(self, team_name, company_name, bid_price):
        if team_name in self.teams and company_name in self.companies:
            team_value = self.teams[team_name]
            company_purse = self.companies[company_name]
            if company_purse >= bid_price:
                self.teams[team_name]["Owner"] = company_name
                self.companies[company_name] -= bid_price
                print(f"{team_name} bought by {company_name} for {bid_price} units.")
            else:
                print(f"{company_name} does not have enough funds to buy {team_name}.")
        else:
            print("Invalid team or company name.")

    def sell_team(self, team_name, company_name):
        if team_name in self.teams and company_name in self.companies:
            team_value = self.teams[team_name]
            if team_value["Owner"] == company_name:
                self.teams[team_name]["Owner"] = "None"
                self.companies[company_name] += team_value["Value"]
                print(f"{team_name} sold by {company_name}.")
            else:
                print(f"{company_name} does not own {team_name}.")
        else:
            print("Invalid team or company name.")

    def trade_teams(self, team1_name, team2_name):
        if team1_name in self.teams and team2_name in self.teams:
            team1_value = self.teams[team1_name]
            team2_value = self.teams[team2_name]
            if team1_value["Owner"] == team2_value["Owner"]:
                print(f"Both teams are owned by the same company ({team1_value['Owner']}).")
            else:
                team1_value["Owner"], team2_value["Owner"] = team2_value["Owner"], team1_value["Owner"]
                print(f"Teams {team1_name} and {team2_name} traded successfully.")
        else:
            print("Invalid team name.")

    def display_purse(self):
        print("Company Purse Details:")
        for company, purse in self.companies.items():
            print(f"{company}: {purse}")

    def display_team_details(self):
        print("Team Details:")
        for team, details in self.teams.items():
            print(f"Team: {team}, Owner: {details['Owner']}, Value: {details['Value']}")

# Example data for teams and companies
teams = {
    "CSK": {"Owner": "None", "Value": 100},
    "MI": {"Owner": "None", "Value": 120},
    "RCB": {"Owner": "None", "Value": 110}
}

companies = {
    "CompanyA": 500,
    "CompanyB": 600,
    "CompanyC": 700
}

trade = Trade(teams, companies)

# Example usage
while True:
    print("\nMenu:")
    print("1. Buy team")
    print("2. Sell team")
    print("3. Trade teams")
    print("4. Display purse")
    print("5. Display team details")
    print("6. Quit")
    choice = input("Enter your choice: ")

    if choice == "1":
        team_name = input("Enter team name: ")
        company_name = input("Enter company name: ")
        bid_price = float(input("Enter bid price: "))
        trade.buy_team(team_name, company_name, bid_price)

    elif choice == "2":
        team_name = input("Enter team name: ")
        company_name = input("Enter company name: ")
        trade.sell_team(team_name, company_name)

    elif choice == "3":
        team1_name = input("Enter first team name: ")
        team2_name = input("Enter second team name: ")
        trade.trade_teams(team1_name, team2_name)

    elif choice == "4":
        trade.display_purse()

    elif choice == "5":
        trade.display_team_details()

    elif choice == "6":
        print("Exiting...")
        break

    else:
        print("Invalid choice.")
```

```python
import random
import matplotlib.pyplot as plt

class Auction_mania:
    def __init__(self):
        self.team_details = {}
        self.team_players = {}
    def load_team_details(self):
        with open("team_details.txt", "r") as file:
            for line in file:
                team, _ = line.strip().split(": ")
                self.team_players[team] = []

    def load_players(self):
        with open("players.txt", "r") as file:
            for line in file:
                team, price = line.strip().split(": ")
                self.team_players[team].append(int(price))

    def show_team_players_chart(self, team):
        plt.bar(range(len(self.team_players[team])), self.team_players[team], color='skyblue')
        plt.xlabel('Player Index')
        plt.ylabel('Purchase Price')
        plt.title(f'Purchase Price of Players for {team}')
        plt.xticks(range(len(self.team_players[team])), [f"Player {i+1}" for i in range(len(self.team_players[team]))], rotation=45)
        plt.tight_layout()
        plt.show()

    def create_team_details_file(self):
        team_details = {
            "Chennai Super Kings": 900000000,
            "Kolkata Knight Riders": 850000000,
            "Delhi Capitals": 750000000,
            "Mumbai Indians": 900000000,
            "Royal Challengers Bangalore": 850000000,
            "Gujarat Titans": 800000000,
            "Rajasthan Royals": 750000000,
            "Punjab Kings": 790000000,
            "Sunrisers Hyderabad": 750000000,
            "Lucknow Super Giants": 850000000
        }

        with open("team_details.txt", "w") as file:
            for team, money in team_details.items():
                file.write(f"{team}: {money}\n")

    def create_players_file(self):
        players = []
        player_names = ["Virat Kohli", "MS Dhoni", "Rohit Sharma", "David Warner", "Kane Williamson", "Joe Root", "Babar Azam", "Steve Smith", "Quinton de Kock", "Shakib Al Hasan",
                        "Kieron Pollard", "Hardik Pandya", "Jasprit Bumrah", "AB de Villiers", "Chris Gayle", "Jonny Bairstow", "Rashid Khan", "Andre Russell", "Shikhar Dhawan", 
                        "Faf du Plessis", "Eoin Morgan", "Ben Stokes", "Ravindra Jadeja", "Mitchell Starc", "Jofra Archer", "Imran Tahir", "Mohammed Shami", "Jos Buttler", 
                        "Yuzvendra Chahal", "Pat Cummins"]

        for name in player_names:
            base_price = random.randint(30000, 70000)
            players.append((name, base_price))

        random.shuffle(players)

        with open("players.txt", "w") as file:
            for player in players:
                file.write(f"{player[0]}: {player[1]}\n")

    def load_team_details(self):
        with open("team_details.txt", "r") as file:
            for line in file:
                team, purse = line.strip().split(": ")
                self.team_details[team] = int(purse)
                self.team_players[team] = []

    def load_players(self):
        players = []
        with open("players.txt", "r") as file:
            for line in file:
                name, base_price = line.strip().split(": ")
                players.append((name, int(base_price)))
        return players

    def buy_players(self, players):
        random_teams = random.sample(list(self.team_details.keys()), 10)
        for team in random_teams:
            if not players:
                print("No more players available for bidding.")
                break
            player = players.pop(0)
            player_name, player_price = player
            if self.team_details[team] >= player_price:
                self.team_details[team] -= player_price
                self.team_players[team].append(player_price)
                print(f"{team} bought {player_name} for {player_price}.")
            else:
                print(f"{team} doesn't have enough money to bid for {player_name}.")

    def update_team_details_file(self):
        with open("team_details.txt", "w") as file:
            for team, purse in self.team_details.items():
                file.write(f"{team}: {purse}\n")

    def update_team_players_files(self):
        for team, players in self.team_players.items():
            with open(f"{team}_players.txt", "w") as file:
                for player in players:
                    file.write(f"{player}\n")

    def show_team_details(self):
        print("\nTeam Details After Auction:")
        for team, purse in self.team_details.items():
            print(f"Team: {team}, Remaining Money: {purse}")

    def show_team_money_charts(self):
        team_names = list(self.team_details.keys())
        team_purses = list(self.team_details.values())

        max_money_index = team_purses.index(max(team_purses))
        max_money_team = team_names[max_money_index]

        # Bar chart
        plt.figure(figsize=(10, 5))
        plt.bar(team_names, team_purses, color='skyblue')
        plt.xlabel('Teams')
        plt.ylabel('Money')
        plt.title('Team Money Before Auction (Bar Chart)')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # Pie chart
        plt.figure(figsize=(7, 7))
        plt.pie(team_purses, labels=team_names, autopct='%1.1f%%', startangle=140)
        plt.title('Team Money Before Auction (Pie Chart)')
        plt.axis('equal')
        plt.show()

        # Plot with highlighted maximum money team
        plt.figure(figsize=(10, 5))
        plt.plot(team_names, team_purses, marker='o', linestyle='-', color='b')
        plt.scatter(max_money_team, self.team_details[max_money_team], color='r', label=f'Max ({max_money_team})')
        plt.xlabel('Teams')
        plt.ylabel('Money')
        plt.title('Team Money Before Auction (Plot)')
        plt.legend()
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    def show_team_players_chart(self, team):
        plt.bar(range(len(self.team_players[team])), self.team_players[team], color='skyblue')
        plt.xlabel('Player Index')
        plt.ylabel('Purchase Price')
        plt.title(f'Purchase Price of Players for {team}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    def show_all_team_players_charts(self):
        for team in self.team_players.keys():
            self.show_team_players_chart(team)

    def run_auction(self):
        self.create_team_details_file()
        self.create_players_file()
        self.load_team_details()
        players = self.load_players()
        random.shuffle(players)

        while True:
            print("\n1. Show all teams' details after auction")
            print("2. Show all teams' money in graphs")
            print("3. Buy players")
            print("4. Show purchase price of players for each team")
            print("5. Quit")

            choice = input("Enter your choice: ")

            if choice == "1":
                self.show_team_details()
            elif choice == "2":
                self.show_team_money_charts()
            elif choice == "3":
                self.buy_players(players)
                self.update_team_details_file()
                self.update_team_players_files()
            elif choice == "4":
                
                self.load_team_details()
                self.load_players()

                while True:
                    print("Choose a team to show player purchase price chart:")
                    for i, team in enumerate(self.team_players.keys()):
                        print(f"{i+1}. {team}")

                    choice = input("Enter your choice (1-10): ")

                    if choice.isdigit() and 1 <= int(choice) <= 10:
                        team_index = int(choice) - 1
                        team_name = list(self.team_players.keys())[team_index]
                        self.show_team_players_chart(team_name)
                        break
                    else:
                        print("Invalid choice. Please enter a number between 1 and 10.")
            elif choice == "5":
                print("Exiting...")
                break
            else:
                print("Invalid choice. Please try again.")

# Create an instance of the Auction_mania class
auction = Auction_mania()
auction.run_auction()
```
